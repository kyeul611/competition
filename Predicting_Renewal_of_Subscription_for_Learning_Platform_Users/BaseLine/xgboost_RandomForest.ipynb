### Data Preprocessing

```python
# user_id 제거
train.pop('user_id')
test_user_id = test.pop('user_id')

# target 분리
target = train.pop('target')
```

```python
# encoding
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
train.head(5)
```

```python
# split data to train_set and valid_set

from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train, target, test_size=0.2, random_state=42)
```

## Random Forest Classifier

```python
# 모델링 (random forest)
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10)
rf.fit(x_train, y_train)
pred_rf = rf.predict(x_valid)
```

```python
from sklearn.metrics import f1_score, log_loss
```

```python
print("f1_score: ", f1_score(y_valid, pred_rf, average='macro'))
# print("log_loss: ", log_loss(y_valid, pred_rf)
```

### Xgboost

```python
# 모델링 (xgboost)
from xgboost import XGBClassifier
xgb = XGBClassifier(random_state=42, n_estimators=600, max_depth=3)
xgb.fit(x_train, y_train)
pred_xgb = xgb.predict(x_valid)
```

```python
print("f1-score: ", f1_score(y_valid, pred_xgb, average='macro'))
```

## Dicision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)
pred_dt = dt.predict(x_valid)
print("f1-score: ", f1_score(y_valid, pred_dt, average='macro'))
```
