### 결측치 처리

시계열 데이터이기 때문에 결측치를 뒤의 값으로 대체하여 처리함.
첫 행에 결측치인 column이 존재하기 때문에 뒤의 값으로 설정.  
모델 성능에 따라 다른 방법을 적용해봐야함.

```python
train_df[train_df['최고기온'].isnull() | train_df['최저기온'].isnull() | train_df['일교차'].isnull()]
```

```python
# 평균기온이 -1.7인 날의 기술 통계값
# 결측치 값을 기술 통계값 중 평균값으로 체우자.
train_df[(train_df['평균기온'] >= -1.7 -1) & (train_df['평균기온'] <= -1.7+1)].describe()
# 최고기온: 2.5
# 최저기온: -5.4
# 일교차: 8
# 강수량: 1.4
# 일사합: 8.7
```

```python
train_df[(train_df['평균기온'] >= 12.3-1) & (train_df['평균기온'] <= 12.3+1) ].describe()

# 최고기온: 17.4
# 최저기온: 7.8
# 일교차: 9.6

```

```python
train_df[(train_df['평균기온'] >= 11.4-1) & (train_df['평균기온'] <= 11.4+1) ].describe()

# 최고기온: 16.4
# 최저기온: 6.9
# 일교차: 9.5
# 강수량: 7.1
# 일조합: 5.9
```

```python
train_df[(train_df['평균기온'] >= 26.8-1) & (train_df['평균기온'] <= 26.8+1) ].describe()

# 최고기온: 30.9
# 최저기온: 23.2
# 일교차: 7.7
# 강수량: 8.5
```

```python
# bfill -> 뒤의 값으로 대체, ffill -> 앞의 값으로 대체
train_df['최고기온'].fillna(method='bfill', inplace=True) 
train_df['최저기온'].fillna(method='bfill', inplace=True)
train_df['일교차'].fillna(method='bfill', inplace=True)
train_df['강수량'].fillna(method='bfill', inplace=True)
train_df['평균풍속'].fillna(method='bfill', inplace=True)
train_df['일조합'].fillna(method='bfill', inplace=True)
train_df['일사합'].fillna(method='bfill', inplace=True)
train_df['일조율'].fillna(method='bfill', inplace=True)
```

### 일시에 따른 평균기온

```python
# '일시' column을 datetime 형식으로 변환
train_df['일시'] = pd.to_datetime(train_df['일시'])

# '년도' 정보를 추출하여 새로운 변수 생성
train_df['년'] = train_df['일시'].dt.year

# 월 변수 생성
train_df['월'] = train_df['일시'].dt.month

# 일 변수 생성
train_df['일'] = train_df['일시'].dt.day

```

```python
# 년도별 그래프 출력
g = sns.FacetGrid(train_df, col='년', col_wrap=4, height=4, sharey=False)
g.map(sns.lineplot, '월', '평균기온', sort=False) 
g.set_axis_labels('Month', 'Average Temperature')
plt.show()
```
